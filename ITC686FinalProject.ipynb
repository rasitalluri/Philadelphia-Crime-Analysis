## Reading data that we need further in the notebook

######Reading Crime Data

```python
file_data = "/FileStore/tables/PhillyCrimeIncidents.csv"

file_schema = StructType([StructField('ObjectID', IntegerType(), True),
                     StructField('DCDist', IntegerType(), True),
                     StructField('PSA', StringType(), True),
                     StructField('DispatchDateTime', StringType(), True),
                     StructField('DispatchDate', StringType(), True),
                     StructField('DispatchTime', StringType(), True),
                     StructField('DispatchHour', StringType(), True),
                     StructField('DCKey', StringType(), True),
                     StructField('LocationBlock', StringType(), True),
                     StructField('UCRGeneral', IntegerType(), True),
                     StructField('TextGeneralCode', StringType(), True),
                     StructField('Point_X', FloatType(), True),
                     StructField('Point_Y', FloatType(), True),
                     StructField('Latitude', FloatType(), True),
                     StructField('Longitude', FloatType(), True)])

crime_df = spark.read.csv(file_data, header=True, schema=file_schema)
```

######Reading Weather Data

```python
weather_data = "/FileStore/tables/PhiladelphiaWeather.csv"

weather_schema = StructType([StructField('DateTime', StringType(), True),
                             StructField('MaxTemperature', IntegerType(), True),
                             StructField('MinTemperature', IntegerType(), True),
                             StructField('Snow', FloatType(), True),
                             StructField('CloudCover', IntegerType(), True),
                             StructField('Precipitation', FloatType(), True)])

weather_df = spark.read.csv(weather_data, header=True, schema=weather_schema)
```

######Reading Districts Shape File

```python
district_pdf = gpd.read_file('https://opendata.arcgis.com/datasets/62ec63afb8824a15953399b1fa819df2_0.zip')
```

## Data Cleaning

```python
# In this step we will format the date and timsteamp column in respective format, normalize the data removing descrepencies
crime_df_new = (crime_df
              .withColumn("DispatchDateAndTime", to_timestamp(col("DispatchDateTime"), "yyyy-MM-dd HH:mm:ss"))
              .withColumn("DispatchMonth", month(to_timestamp(col("DispatchDateTime"), "yyyy-MM-dd HH:mm:ss")))
              .withColumn("DispatchMonthName", date_format(to_timestamp(col("DispatchDateTime"), "yyyy-MM-dd HH:mm:ss"),'MMM'))
              .withColumn("DispatchYear", year(to_timestamp(col("DispatchDateTime"), "yyyy-MM-dd HH:mm:ss")))
              .withColumn("DispatchDate", to_date(col("DispatchDate"), "yyyy-MM-dd"))
              .withColumn("DispatchHour",hour(to_timestamp(col("DispatchDateTime"), "yyyy-MM-dd HH:mm:ss")))
              .withColumn('DispatchDay', date_format(col("DispatchDateTime"), "EEEE"))
              .withColumn('DispatchDayNumber', (dayofweek(col("DispatchDateTime"))))
              .withColumn("DCKey",col("DCKey").cast(LongType()))
              .withColumn("CrimeType", when(col('UCRGeneral') < 800, "Voilent").otherwise("Non-Voilent"))
              .withColumn("CrimeDescription", when(col('UCRGeneral') == 100, "Homicide").when(col('UCRGeneral') == 300, "Robbery")
                          .when(col('UCRGeneral') == 400, "Aggravated Assault")
                          .when(col('UCRGeneral') == 500, "Burglary")
                          .when(col('UCRGeneral') == 600, "Theft")
                          .when(col('UCRGeneral') == 700, "Motor Vehicle Theft").otherwise(col("TextGeneralCode")))
              .withColumn("EndPos", when(instr('LocationBlock',' ST')!=0, instr('LocationBlock',' ST')+3)
                          .when(instr('LocationBlock',' RD')!=0,instr('LocationBlock',' RD')+3)
                          .when(instr('LocationBlock',' AVE')!=0,instr('LocationBlock',' AVE')+4)
                          .when(instr('LocationBlock',' AV')!=0,instr('LocationBlock',' AV')+3)
                          .when(instr('LocationBlock',' WAY')!=0,instr('LocationBlock',' WAY')+4)
                          .when(instr('LocationBlock',' BLVD')!=0,instr('LocationBlock',' BLVD')+5)
                          .when(instr('LocationBlock',' LN')!=0,instr('LocationBlock',' LN')+3)
                          .when(instr('LocationBlock',' LA')!=0,instr('LocationBlock',' LA')+3)
                          .when(instr('LocationBlock',' DR')!=0,instr('LocationBlock',' DR')+3)
                          .otherwise(length('LocationBlock')))
                .withColumn("StartPos", when(instr('LocationBlock','BLOCK')!=0, instr('LocationBlock','BLOCK') + 6)
                          .otherwise(0))
                .withColumn("StreetVal", expr("""substring(LocationBlock,StartPos,EndPos - StartPos)"""))
                .withColumn("Street", when((expr("""length(StreetVal)< 2 """)),col('LocationBlock')).otherwise(col('StreetVal')))
                .drop("Point_X")
                .drop("Point_Y")
                .drop("EndPos")
                .drop("StartPos")
                .drop("DispatchDateTime")
                .drop("StreetVal")
               )
```

```python
# Checking if there are any NULLs in columns

crime_df_new.cache()
print("Number of rows: ", crime_df_new.count())
print("Number of NULL in DCDist: ", crime_df_new.where(crime_df_new["DCDist"].isNull()).count())
print("Number of NULL in DispatchDate: ", crime_df_new.where(crime_df_new["DispatchDate"].isNull()).count())
print("Number of NULL in DispatchHour: ", crime_df_new.where(crime_df_new["DispatchHour"].isNull()).count())
print("Number of NULL in UCRGeneral: ", crime_df_new.where(crime_df_new["UCRGeneral"].isNull()).count())
print("Number of NULL in TextGeneralCode: ", crime_df_new.where(crime_df_new["TextGeneralCode"].isNull()).count())
print("Number of NULL in LocationBlock: ", crime_df_new.where(crime_df_new["LocationBlock"].isNull()).count())
print("Number of NULL in Latitude: ", crime_df_new.where(crime_df_new["Latitude"].isNull()).count())
print("Number of NULL in Longitude: ", crime_df_new.where(crime_df_new["Longitude"].isNull()).count())
print("Number of NULL in PSA: ", crime_df_new.where(crime_df_new["PSA"].isNull()).count())
```

```python
# Drop the records for which the data is missing
crime_df_dropnull = (crime_df_new
              .where(col("LocationBlock").isNotNull())
              .where(col("Latitude").isNotNull())
              .where(col("Longitude").isNotNull())
              .where(col("PSA").isNotNull())
                     .where(col("DispatchDate")<=to_date(lit("2021-06-30"),"yyyy-MM-dd"))
                    )

crime_df_new.unpersist()

# Removing the duplicates to eliminate the calls reported by multiple people about the same incident
crime_df_dedup = crime_df_dropnull.dropDuplicates(['DispatchDateAndTime','Latitude','Longitude','LocationBlock','UCRGeneral','TextGeneralCode'])

# Remove the records with lat longs outside Philadelphia
crime_df_clean = (crime_df_dedup
              .where(col("Latitude") >= 39.87502))
```

```python
# Format weather data so that it can be merged with crime dataset
weather_df_new = (weather_df
                  .withColumn("YearPart", split(col("DateTime"),'/')[2])
                  .withColumn("TimePart", split(col("YearPart"),' ')[1])
                  .withColumn("Year", split(col("YearPart"),' ')[0])
                  .withColumn("Day", split(col("DateTime"),'/')[1])
                  .withColumn("Month", split(col("DateTime"),'/')[0])
                  .withColumn("MonthVal", when(length("Month")==1, concat(lit("0"),col("Month")))
                              .otherwise(col("Month")))
                  .withColumn("DayVal", when(length("Day")==1, concat(lit("0"),col("Day")))
                              .otherwise(col("Day")))
                  .withColumn("DatePart", concat_ws('-',col("Year"),col("MonthVal"),col("DayVal")))
                  .withColumn("Date", to_date(col("DatePart"),"yyyy-MM-dd"))
                  .withColumn("Hour", split(col("TimePart"),':')[0])
                  .drop("YearPart")
                  .drop("TimePart")
                  .drop("Year")
                  .drop("Day")
                  .drop("DayVal")
                  .drop("Month")
                  .drop("MonthVal")
                  .drop("DateTime")
                  .drop("DatePart")
                 )
```

```python
# Join Crime and weather dataset based on date and hour, The weather data that we are using is hourly dataset
crime_df_data = crime_df_clean.join(weather_df_new, ((crime_df_clean['DispatchDate'] == weather_df_new['Date']) & (crime_df_clean['DispatchHour'] == weather_df_new['Hour']))).drop("Date").drop("Hour")

# Partition the dataset based on Year
crime_df_data.repartition("DispatchYear")

# We will be using this dataset very frequently in further part of the notebook, caching this dataframe
crime_df_data.cache()
```

## Classification and crime probability Prediction

```python
# Importing the required packages classification
from pyspark.ml.classification import RandomForestClassifier, LogisticRegression, DecisionTreeClassifier, GBTClassifier
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder, StandardScaler, ChiSqSelector
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator,BinaryClassificationEvaluator
```

```python
# Preparing the data for classification
crime_data_for_pred_1 = (crime_df_data              .select("DispatchDate","DCDist","PSA","DispatchHour","DispatchDayNumber","DispatchYear","DispatchMonth","Latitude","Longitude","MinTemperature","Snow","CloudCover","Precipitation",expr("CrimeDescription AS CrimeType")))

crime_data_for_pred_2 = (crime_data_for_pred_1
                         .withColumn("SnowyDay", ceil(crime_data_for_pred_1['Snow']))
                         .withColumn("CloudCoverage", when(col('CloudCover').isin(0,100),col('CloudCover')).otherwise(50))
                         .withColumn("RainyDay", ceil(crime_data_for_pred_1['Precipitation']))
                         .withColumn('DateNumber', datediff(crime_data_for_pred_1['DispatchDate'], to_date(lit('2010-01-01'), format='yyyy-MM-dd')))
                         .withColumn('MonthDay', dayofmonth(crime_data_for_pred_1['DispatchDate']))
                         .drop("Snow")
                         .drop("CloudCover")
                         .drop("Precipitation")
                         .drop("DispatchDate")
                        )
```

```python
# String Indexing categorical variables and label
PSAIndexer = StringIndexer(inputCol='PSA',outputCol='PSAIndex')
PSAModel = PSAIndexer.fit(crime_data_for_pred_2)
crime_data_for_pred_2 = PSAModel.transform(crime_data_for_pred_2)

CrimeTypeIndexer = StringIndexer(inputCol='CrimeType',outputCol='CrimeTypeIndex')
CrimeTypeModel = CrimeTypeIndexer.fit(crime_data_for_pred_2)
crime_data_for_pred_2 = CrimeTypeModel.transform(crime_data_for_pred_2)
```

```python
# Onehot Encoding
PSAEncoder = OneHotEncoder(inputCol='PSAIndex',outputCol='PSAEncoded')
PSAEncoderModel = PSAEncoder.fit(crime_data_for_pred_2)
crime_data_for_pred_2 = PSAEncoderModel.transform(crime_data_for_pred_2)
```

```python
# Creating vector for features
IndexedFeatures = ['DCDist','PSAEncoded','DispatchHour','DispatchDayNumber','DispatchYear','DispatchMonth','MonthDay','MinTemperature','SnowyDay','CloudCoverage','RainyDay']

assembler = VectorAssembler(inputCols=IndexedFeatures,outputCol='AssembledFeatures')

crime_data_for_pred_2 = assembler.transform(crime_data_for_pred_2)

display(crime_data_for_pred_2.limit(10))
```

```python
# Scaling features to improve performance
FeatureScaler = StandardScaler(inputCol="AssembledFeatures",outputCol="features")
FeatureScalerModel = FeatureScaler.fit(crime_data_for_pred_2)
crime_data_for_prediction = FeatureScalerModel.transform(crime_data_for_pred_2)
```

```python
train_df, test_df = crime_data_for_prediction.randomSplit([0.7, 0.3], seed=42)
train_df = train_df.repartition("DispatchYear")
train_df.cache()
```

###### Logistic Regression Classification

###### Decision Tree Classification

```python
crime_dt = DecisionTreeClassifier(featuresCol='features',labelCol='CrimeTypeIndex',maxDepth=15)
crime_dt_model = crime_dt.fit(train_df)

crime_pred_dt = crime_dt_model.transform(test_df)
evaluator = MulticlassClassificationEvaluator(
    labelCol="CrimeTypeIndex", predictionCol="prediction", metricName="accuracy")

accuracy_dt = evaluator.evaluate(crime_pred_dt)
print(accuracy_dt)
```

###### Random Forest Classification

```python
crime_rf = RandomForestClassifier(featuresCol='features',labelCol='CrimeTypeIndex', numTrees=15)
crime_rf_model = crime_rf.fit(train_df)

crime_pred_rf = crime_rf_model.transform(test_df)
evaluator = MulticlassClassificationEvaluator(
    labelCol="CrimeTypeIndex", predictionCol="prediction", metricName="accuracy")

accuracy_rf = evaluator.evaluate(crime_pred_rf)
print(accuracy_rf)
```

## Binary Classification

Since the Multi Classification models are not giving expected results we tried implementing binary classification

We are only concentrating on Thefts, The data is generated with label 1 for the combination of features where the theft is reported and the remaining are set to 0. We applied multiple binary classification methods. The results are much betters and "Gradient Boosting Classifier" has outperformed other models. 

Problem Statement: Predicting the possibility of theft based on given feature combination.
Models Used: Logistic Regression, Gradient Boosting Classification, Random Forest Classification

Performance:
Logistic Regression --- Area Under ROC 61%, Accuracy: 58.9%
Gradient Boosting Classifier --- Area Under ROC 76.3%
Random Forest Classifier --- Area Under ROC 71%

Output: Gradient boosting classifier has performed better than other models.

```python
# Preparing Data For Binary Classification
weather_df_new.registerTempTable("df_weatherfeatures")

crime_df_data.registerTempTable("df_crimedata")

spark.sql("SELECT DCDist FROM df_crimedata GROUP BY DCDist").registerTempTable("df_districts")

# Selecting all the required features
spark.sql("SELECT dw.Date,cast(dw.Hour as int) Hour,dd.DCDist,dayofmonth(dw.Date) DayOfMonth,year(dw.Date) Year,dayofweek(dw.Date) DayOfWeek,month(dw.Date) Month, \
dw.MinTemperature,CASE WHEN dw.Snow = 0 THEN 0 ELSE 1 END AS SnowyDay, CASE WHEN dw.CloudCover IN (0,100) THEN dw.CloudCover ELSE 50 END AS CloudCoverage, \
CASE WHEN dw.Precipitation = 0 THEN 0 ELSE 1 END AS RainyDay \
FROM df_weatherfeatures as dw CROSS JOIN df_districts as dd WHERE dw.Date <= to_date('2021-06-30','yyyy-MM-dd')").registerTempTable("df_timefeatures")

# Merge with Crime data set and set the label for theft
df_data_theft = spark.sql("SELECT dt.Date,dt.Hour,dt.DCDist,dt.DayOfMonth,dt.Year,dt.DayOfWeek,dt.Month,dt.MinTemperature,dt.SnowyDay,dt.CloudCoverage,dt.RainyDay, \
DOUBLE(MAX(CASE WHEN ct.UCRGeneral = 600 THEN 1 ELSE 0 END)) as label \
FROM df_timefeatures as dt LEFT JOIN df_crimedata ct ON (dt.Date = ct.DispatchDate AND dt.Hour = ct.DispatchHour AND dt.DCDist = ct.DCDist AND ct.UCRGeneral = 600) \
GROUP BY dt.Date,dt.Hour,dt.DCDist,dt.DayOfMonth,dt.Year,dt.DayOfWeek,dt.Month,dt.MinTemperature,dt.SnowyDay,dt.CloudCoverage,dt.RainyDay")
```

```python
# Feature vecorization
IndexedFeatures = ['DCDist','Year','Month','Hour','DayOfMonth','DayOfWeek','MinTemperature','SnowyDay','CloudCoverage','RainyDay']
assembler = VectorAssembler(inputCols=IndexedFeatures,outputCol='AssemblerFeatures')
df_data_theft = assembler.transform(df_data_theft)
display(df_data_theft.limit(10))
```

```python
# Applying Standard Scalar to improve the performance
FeatureScaler = StandardScaler(inputCol="AssemblerFeatures",outputCol="ScaledFeatures")
FeatureScalerModel = FeatureScaler.fit(df_data_theft)
df_data_theft = FeatureScalerModel.transform(df_data_theft)
```

```python
# Train and test Split
train_data_df, test_data_df  = df_data_theft.randomSplit([0.8, 0.2], seed=12345)
train_data_df = train_data_df.repartition("Year","DCDist")
train_data_df.cache()
```

###### We are checking the weights of both the classes, since the number of rows with 1 is only 14% of the entire dataset, we will be assigning the weights to each class to increase the performace of the models

```python
DatasetSize = float(train_data_df.select("label").count())
NumPos = train_data_df.select("label").where('label == 1').count()
PercentOnes = (float(NumPos)/float(DatasetSize))*100
NumNeg = float(DatasetSize-NumPos)
print('The number of ones are {}'.format(NumPos))
print('Percentage of ones are {}'.format(PercentOnes))
```

```python
BalanceRatio = NumNeg/DatasetSize
train_data_df = train_data_df.withColumn("ClassWeight", when(train_data_df.label == 1,BalanceRatio).otherwise(1-BalanceRatio))
train_data_df.select("label","ClassWeight").show(5)
```

```python
ChisqSelect = ChiSqSelector(featuresCol='ScaledFeatures',outputCol='Aspect',labelCol='label',fpr=0.05)
train_data_df=ChisqSelect.fit(train_data_df).transform(train_data_df)
test_data_df=ChisqSelect.fit(test_data_df).transform(test_data_df)
```

###### Logistic Regression

```python
TheftCrimeLR = LogisticRegression(labelCol="label", featuresCol="Aspect",weightCol="ClassWeight",maxIter=10)
TheftCrimeLRModel = TheftCrimeLR.fit(train_data_df)
theft_predict_lr=TheftCrimeLRModel.transform(test_data_df)
theft_predict_lr.select("AssemblerFeatures","label","prediction","probability").show(10)
```

```python
evaluator = BinaryClassificationEvaluator(rawPredictionCol='rawPrediction',labelCol='label')
print("The area under ROC for test set using logistic regression is {}".format(evaluator.evaluate(theft_predict_lr)))
print("Accuracy for test set using logistic regression is {}".format(TheftCrimeLRModel.summary.accuracy))
```

###### Gradient Booster Classification

```python
TheftCrimeGBT = GBTClassifier(featuresCol='ScaledFeatures',labelCol='label',weightCol="ClassWeight")
TheftCrimeGBTModel = TheftCrimeGBT.fit(train_data_df)
theft_predict_gbt_train = TheftCrimeGBTModel.transform(train_data_df)
theft_predict_gbt = TheftCrimeGBTModel.transform(test_data_df)
theft_predict_gbt.select("AssemblerFeatures","label","prediction").show(10)
```

```python
print("The area under ROC for train set using gradient boosting classifier is {}".format(evaluator.evaluate(theft_predict_gbt_train)))
print("The area under ROC for test set using gradient boosting classifier is {}".format(evaluator.evaluate(theft_predict_gbt)))
```
